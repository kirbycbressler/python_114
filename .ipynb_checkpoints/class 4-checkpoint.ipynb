```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

class DLList:
    def __init__(self, data):
        self.data = data
        self.head = None
        self.tail = None
    
    def insert(self, data, index):
        if (index > self.count) | (index < 0):
            raise ValueError("Index out of range")
        
    def traverse(self):
        current_node = self.head
        while current_node != None:
            return current_node.data
            
        current_node = current_node.next
    
    def get_list_size(self):
        
        count = 0
        
        current_node = self.head
        
        while current_node != None:
            count += 1
            current_node = current_node.next
            
        return count
    
    def push(self, data):
        # def new node
        
        new_node = Node(data)
        
        new_node.next_node = self.head
        
        new_node.prev_node = None
        
        if self.head != None:
            self.head.prev = new_node
        
        self.head = new_node
        
    
    def append(self, data):  #append data to end of a dllist.
        if not self.head:
            new_node = DNode(data)
            self.head = new_node
        else:
            current = self.head 
            while current:
                current = current.next
            new_node = DNode(data)
            current.next = new_node
            new_node.prev = current
    
    def remove_last(self, data):
        current = self.head
        if not self.head:
            return False
        if current.data == data:
            self.head = self.head.next
            self.head.prev = None
            return True
        
    
    def index(self, data):
        start = self.head
        for i in range(self.count):
            if(start.data == data):
                return i
            start = start.next
        return None
    

# new list
double_list = DLList(4)

#insert
double_list.append(100)
print(double_list)
```
